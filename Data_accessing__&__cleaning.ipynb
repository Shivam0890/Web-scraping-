# Data Quality Dimensions

Completeness --> is data missing ?

Validity  --> is data invalid ---> negative height , duplicated

Accuracy --> data is valid but not acccurate ---> weight 1kg according to data senario

Consistency ---> both valid and accurate but written differently -
e.g. New Youk  & NY




## Order of severity

completeness <--- validity <--- accuray <--- consistency
